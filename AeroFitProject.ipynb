# Statistics from above data

1. Age - The age range of customers spans from 18 to 50 year, with an average age of 29 years.

2. Education - Customer education levels vary between 12 and 21 years, with an average education duration of 16 years.

3. Usage - Customers intend to utilize the product anywhere from 2 to 7 times per week, with an average usage frequency of 3 times per week.

4. Fitness - On average, customers have rated their fitness at 3 on a 5-point scale, reflecting a moderate level of fitness.

5. Income - The annual income of customers falls within the range of USD 30,000 to USD 100,000, with an average income of approximately USD 54,000.

6. Miles - Customers' weekly running goals range from 21 to 360 miles, with an average target of 103 miles per week.

```python
palette_color = sns.color_palette('bright')
title = ['count_of_product', 'count_of_gender', 'count_of_marital_status', 'count_of_fitness']
for i, col in enumerate(['Product', 'Gender', 'MaritalStatus', 'Fitness']):
    count_of_product = data[col].dropna().value_counts().reset_index()
    data_count = list(count_of_product[col])
    keys = list(count_of_product.index)
    plt.pie(data_count, labels=keys, colors=palette_color, autopct='%.0f%%')
    plt.title(title[i])
    plt.show()
    print('-'*100)
```

# Probabilty of Purchasing a Product based on Gender

```python
sns.heatmap(pd.crosstab(data['Product'], data['Gender'], normalize=True, margins=True), annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Gender')
plt.tight_layout()
plt.show()
```

# Insights

1. Probability of buying a treadmill is greater in male than females.
2. Probability of buying KP281 is same in male and females.
3. Females are least likely to buy KP781.

# Probability of Purchasing A Product Based On Age Group

```python
bins = [0, 15, 25, 35, 45, 55, 65, np.inf]
labels = ['0-15','15-25', '25-35', '35-45', '45-55', '55-65', '65+']  # Adjusted labels
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels)
pva = pd.crosstab(data['Product'], data['Age_Group'], normalize=True, margins=True)
pva
```

```python
sns.heatmap(pva, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Age Group')
plt.tight_layout()
plt.show()
```

# Insights

1. Age Group has no significant relation with product buying.

# Probability of Purchasing a Product Based on the Usage.

```python
pvu = pd.crosstab(data['Product'], data['Usage'], normalize=True, margins=True)
pvu
```

```python
sns.heatmap(pvu, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Usage')
plt.tight_layout()
plt.show()
```

# Insights

# Probability of buying a product based on income

```python
bins = [25000, 50000, 75000, 100000, np.inf]
labels = ['$25000-$50000', '$50000-$75000', '$75000-$100000', '$100000+']
data['IncomeRange'] = pd.cut(data['Income'], bins=bins, labels=labels)
pva = pd.crosstab(data['Product'], data['IncomeRange'], normalize=True, margins=True)
pva
pvi = pd.crosstab(data['Product'], data['IncomeRange'], normalize=True, margins=True)
pvi
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(pvi, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Fitness')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# Insights

1.Customers earning between $25k−$75k are most likely to purchase a treadmill.
